```python
# Import required libraries
import os
import pandas as pd
from pathlib import Path
import hvplot.pandas
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# Import required Data,including Increase and Decrease
date = '09-13-2022'
increase = pd.read_csv(os.path.join(Path.cwd(),f'../Data/Increase/stocks-increase-change-in-open-interest-{date}.csv'))
decrease = pd.read_csv(os.path.join(Path.cwd(),f'../Data/Decrease/stocks-decrease-change-in-open-interest-{date}.csv'))

#show dataframe

display(increase.head())
display(decrease.head())
```

```python
# check Missing Data

display(increase.isnull().sum())
display(increase.isnull().sum())
```

```python
# Handle Missing Data

increase = increase.dropna()
decrease = decrease.dropna()
```

```python
#Select the data only equal to minimun DTE (closest to expired).
increase_min_DTE = increase[increase['DTE'] == increase['DTE'].min()]
decrease_min_DTE = decrease[decrease['DTE'] == decrease['DTE'].min()]

# show dataframe
display(increase_min_DTE)
display(decrease_min_DTE)
```

```python
#Merge data of increase and decrease dataframe

combine_df = increase_min_DTE.merge(decrease_min_DTE, how= 'outer')

```

```python
# Sort by this orders Symbol,Type,Strike,Open Int,Volume,OI Chg,IV

combine_sort_df = combine_df.sort_values(['Symbol','Type','Strike','Open Int','Volume','OI Chg','IV'])
```

```python
# Create the function for clean data acquire.
def get_data(date):

        increase = pd.read_csv(os.path.join(Path.cwd(),f'../Data/Increase/stocks-increase-change-in-open-interest-{date}.csv'))
        decrease = pd.read_csv(os.path.join(Path.cwd(),f'../Data/Decrease/stocks-decrease-change-in-open-interest-{date}.csv'))
        combine_df = increase.merge(decrease, how= 'outer')
        combine_min_DTE = combine_df[combine_df['DTE'] == combine_df['DTE'].min()]
        combine_min_DTE = combine_min_DTE.dropna()
        combine_sort_df = combine_min_DTE.sort_values(['Symbol','Type','Strike','Open Int','Volume','OI Chg','IV'])
        
        return combine_sort_df
        
```
